```python
# Import Dependencies
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
```

```python
csvpath_2019 = "Possible Datasets/2019/2019_AGGREGATED.csv"
csvpath_2020 = "Possible Datasets/2020/Hawaii_flights_2020.csv"
csvpath_2021 = "Possible Datasets/2021/Hawaii_flights_2021.csv"

flights_2019 = pd.read_csv(csvpath_2019)
flights_2020 = pd.read_csv(csvpath_2020)
flights_2021 = pd.read_csv(csvpath_2021)
```

```python
len(flights_2019)
```

```python
flights_2019.head()
```

```python
flights_2019 = flights_2019[flights_2019["DEST_STATE_NM"]== "Hawaii"]
```

```python
len(flights_2019)
```

```python
flights_2019 = flights_2019[flights_2019["ORIGIN_STATE_NM"]!= "Hawaii"]
flights_2020 = flights_2020[flights_2020["ORIGIN_STATE_NM"]!= "Hawaii"]
flights_2021 = flights_2021[flights_2021["ORIGIN_STATE_NM"]!= "Hawaii"]
```

```python
len(flights_2019)
#len(flights_2020)
#len(flights_2021)
```

```python
hawaii_flights = pd.concat([flights_2019,flights_2020,flights_2021])
```

```python
hawaii_flights
```

```python
clean_data_df= hawaii_flights.rename(columns=
                                
                                         {"YEAR": "Year",
                                         "QUARTER":"Quarter",
                                         "ORIGIN":"Origin",
                                         "ORIGIN_STATE_ABR":"State departure",
                                         "DEST":"Destination Airport",
                                         "DEST_STATE_ABR":"Destination State",
                                         "AIRPORT_GROUP":"Airport Group",
                                         "TICKET_CARRIER":"Ticket Carrier",
                                         "PASSENGERS":"Total Passengers",
                                         "MARKET_FARE":"Ticket Cost"
                                        }
                                    )

clean_data_df
```

```python
airports_data_df = clean_data_df[['Year', 'Quarter','Origin','State departure','Destination Airport','Airport Group',
                                       'Ticket Carrier','Total Passengers','Ticket Cost']]

airports_data_df
```

```python
#JE Viz questions to explore
    # "Show most popular orgin"
    # "Passenger count year over year"
    
#report columns to pull from in code
airports_data_df.columns
```

```python
#JE Viz questions to explore
    # "Show most popular orgin"

#identify count of passengers per origin for each year and 
    #sort by top passengers descending to show top 5 origins with the most 
    #passengers flying into hawaii

top_origin = airports_data_df[["Ticket Carrier","Total Passengers", "Ticket Cost"]].groupby("Ticket Carrier").sum().sort_values(by="Ticket Cost", ascending=False)

#airports_data_df.groupby(airports_data_df["Ticket Carrier"]).sum().sort_values(by="Total Passengers", ascending=False)

top_origin.plot.bar()

#same as above but broken down by year
```

```python
#JE Viz questions to explore
    # "Passenger count year over year"

#narrow down dataframe to exactly what variables we want to see
passenger_count_by_year_df = airports_data_df[["Total Passengers", "Ticket Cost", "Year"]].groupby("Year").sum()

passenger_count_by_year_df

#plot passenger count by year
passenger_count_by_year_df.plot.bar(y="Ticket Cost")
```
